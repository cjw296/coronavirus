```python
from bars import plot_bars
from heatmap import heatmap
from phe import demographic_stream_plot, load_demographic_data, diff
```

```python
data = load_demographic_data('case_demographics_nation', 'England', 'rollingRate', start='2021-09-11')[0]
```

```python
heatmap(data, title='Rolling rate of new cases')
```

```python
heatmap(diff(data, days=7), title='Difference for rolling rate of new cases between T and T-7', 
        figsize=(16, 9), center=0, cmap='coolwarm')
```
