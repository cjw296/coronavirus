```python
import numpy as np
import pandas as pd
from plotting import stacked_area_plot
from matplotlib import pyplot as plt
```

```python
a = np.linspace(0, 2*np.pi, 1000)
data = pd.DataFrame({'sin': np.sin(a), '-sin': -np.sin(a), 'cos': np.cos(a)}, index=a)
```

```python
fig, ax = plt.subplots(nrows=2, figsize=(16, 9))
data.plot(ax=ax[0])
stacked_area_plot(ax[-1], [data['sin'], data['-sin'], data['cos']])
```

```python
fig, ax = plt.subplots(ncols=1, figsize=(16, 9))
stacked_area_plot(ax, [data['sin'], data['-sin'], data['cos']], vertical=True)
```

```python
fig, ax = plt.subplots(ncols=3, figsize=(16, 9), sharex=True, sharey=True)
data_ = data.sort_index(ascending=True)
stacked_area_plot(ax[0], [data['sin']], vertical=True)
stacked_area_plot(ax[1], [data['sin'], data['-sin']], vertical=True)
stacked_area_plot(ax[2], [data['sin'], data['-sin'], data['cos']], vertical=True)
ax[0].invert_yaxis()
```
